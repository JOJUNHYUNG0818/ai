```python
# inner/outer
index1 = [
    datetime(2022, 9, 8),
    datetime(2022, 9, 7)
]

index2 = [
    datetime(2022, 9, 6),
    datetime(2022, 9, 5)
]
df1 = pd.DataFrame(np.arange(0, 6).reshape(2,3), index=index1, columns=['open', 'high', 'low'])
df2 = pd.DataFrame(np.arange(6,12).reshape(2,3), index=index2, columns=['high', 'low', 'close'])
df3=pd.concat([df1,df2], join='outer') # 전부 출력 없는부분 NaN값
df3
df4=pd.concat([df1,df2], join='inner') # 공통 열만 출력
df3
```

```python
# groupby
data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]

columns = ["테마", "종목명", "PER", "PBR"]
df = pd.DataFrame(data=data, columns=columns)
df.groupby(by='테마').mean()
df.groupby(by='종목명').sum()
# 위 데이터프레임에서 테마별 종목의 개수를 계산하세요.
df.groupby('테마').size()
# 12. 위 데이터프레임을 '테마' 컬럼으로 그룹핑한 후 '2차전지(생산)' 그룹만 가져오세요.
gb1=df.groupby(by='테마')
gb1.get_group('2차전지(생산)')
# 위 데이터프레임에서 테마별 PER의 평균을 계산하세요.
df.groupby(by='테마').mean()['PER']
# 위 데이터프레임에 대해 테마로 그룹화한 후 PER, PBR 컬럼을 슬라이싱하세요. 그 다음 두 컬럼에 대한 평균을 계산하세요.
df.groupby(by='테마').agg({'PER':max,'PBR':min})
df1=df.groupby(by='테마')['PER','PBR'].agg([np.mean,np.std])
df1
```

```python
df = sns.load_dataset( 'titanic' )
df['who'].value_counts()
def transform_who(x):
    if x=='man':
      return '남자'
    elif x == 'woman':
      return '여자'
    else :
      return '아이'

df['who'].apply(transform_who).value_counts()


# 좌석요금 : fare
def tr_fare_age(x):
  return x['fare']/x['age']

df.apply(tr_fare_age,axis=1)
df.groupby(by='sex')['survived'].mean()
df.groupby(by=['sex','pclass']).mean()
df.groupby(by=['sex','pclass'])['survived'].mean()
df.groupby(by=['sex','pclass'])['survived'].mean().reset_index()
df.groupby(by=['sex','pclass']).agg({'survived':py.mean, 'age':py.sum})
```
