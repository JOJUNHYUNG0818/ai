```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train[1] # = 'truck'
cifar10_label = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
#cifar10_label[6]
cifar10_label[y_train[1][0]]
```

```python
import matplotlib.pyplot as plt
cifar10_label = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]

rows = 3
cols = 5
j=0
fig=plt.figure()
for i in range(1,rows*cols+1):
  ax=fig.add_subplot(rows,cols,i)
  ax.imshow(x_train[j])
  ax.set_xlabel(cifar10_label[y_train[j][0]])  # 19번줄확인
  ax.set_xticks([]) # x축눈금제거
  ax.set_yticks([])
  j+=1
plt.show()
```
