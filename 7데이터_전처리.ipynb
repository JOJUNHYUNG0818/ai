```python
import cv2 as cv
import matplotlib.pyplot as plt
import random

def random_crop(img, minH=32, minW=32,H=170,W=280):
  img_height,img_width = img.shape[:2] # 실제 이미지의 높이 너비를 넣어준다
  randomH=random.randrange(minH,H)
  randomW=random.randrange(minW,W)

  random_y=random.randrange(0,img_height-randomH)
  random_x=random.randrange(0,img_width-randomW)

  crop_img=img[random_y:random_y+randomH, random_x:random_x+randomW]

  resized_img = cv.resize(crop_img,(H,W),interpolation=cv.INTER_LINEAR) # 사이즈를 바꿨기 때문에 화소가 약해짐
  return resized_img

```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import random
import numpy as np

def add_gaussian_noise(img,std):
  img_height, img_width = img.shape[:2]
  noise_img = np.zeros((img_height,img_width,3),dtype=np.float32)# 이미지 크기 만큼 2차원 0으로 채움
  for j in range (img_height):
    for i in range (img_width):
      noise = np.random.normal();
      noise_img[j][i] = img[j][i]+(noise*std)
  return np.asarray(noise_img,dtype=np.uint8)

```

```python
img_g = add_gaussian_noise(img,30)
img_g = add_gaussian_noise(img,20)
img_g = add_gaussian_noise(img,40)
img_g
```

```python
#Gaussian Blur
import cv2 as cv
import matplotlib.pyplot as plt

def add_gaussian_blur(img,ksize=(5,5),sigma=10):
  blur_img = cv.GaussianBlur(img,ksize,sigma)
  return blur_img

blur_img_r = add_gaussian_blur(img)
blur_img_r
```
